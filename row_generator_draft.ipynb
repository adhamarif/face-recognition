```python
names = ["dennis","justin","kenneth","kenneth_brille","liza","liza_brille","miriam"]
# root = os.getcwd()
image_folder = "C:\\Users\\ASUS\\datasets\\image"
faces_folder = "C:\\Users\\ASUS\\datasets\\face"
```

Creating dataframe by row

```python
def row_generator(folder_list,names_list):

    for folder in folder_list:
        for name in names_list:
            sub_dir = os.path.join(folder,name)
            if os.path.exists(sub_dir):
                for filename in os.listdir(sub_dir):
                    file_path = os.path.join(sub_dir,filename)
                    yield file_path,name

row_gen = row_generator([image_folder,faces_folder],names)
# next(row_gen)

image_counter = 0
try:
    while True:
        # Attempt to get the next item
        next_item = next(row_gen)
        image_counter += 1
        print(next_item)
except StopIteration:
    print(f"End of iterator reached with {image_counter} images.")
```
