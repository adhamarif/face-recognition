Importing libraries

```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
import csv
```

```python
#names = ["dennis","justin","kenneth","kenneth_brille","liza","liza_brille","miriam"]
names = ['steffen', 'adham']
root = os.getcwd()
image_folder = r"C:\Users\Asus\Desktop\Project\face-recognition\datasets\image"
faces_folder = r"C:\Users\Asus\Desktop\Project\face-recognition\datasets\face"
```

Creating dataframe by row

```python
def row_generator(folder_list,names_list):

    for folder in folder_list:
        for name in names_list:
            sub_dir = os.path.join(folder,name)
            if os.path.exists(sub_dir):
                for filename in os.listdir(sub_dir):
                    file_path = os.path.join(sub_dir,filename)
                    yield file_path,name

row_gen = row_generator([image_folder,faces_folder],names)
# next(row_gen)

image_counter = 0
try:
    while True:
        # Attempt to get the next item
        next_item = next(row_gen)
        image_counter += 1
        print(next_item)
except StopIteration:
    print(f"End of iterator reached with {image_counter} images.")
```

Creating dataframe by column

```python
# Specify the path of the CSV file
csv_file_path = os.path.join(root, 'datasets/label.csv')

# Open the CSV file in write mode
with open(csv_file_path, 'w', newline='') as csv_file:
    # Create a CSV writer
    csv_writer = csv.writer(csv_file)

    # Write the header to the CSV file
    csv_writer.writerow(['filepath', 'label'])

    # Iterate over the rows generated by 'row_generator'
    row_gen = row_generator([image_folder], names)
    for image in tqdm(row_gen):
        # Write each row to the CSV file
        csv_writer.writerow([image[0], image[1]])
    
```
