3. One-Hot encode your label column.

````python
data_df_encoded = pd.get_dummies(data_df,columns=["label"],dtype=float)
data_df_encoded
````

````python
data_df_encoded = data_df_encoded.drop(["id"],axis=1)
data_df_encoded
````

````python
data_df_encoded["label"] = data_df["label"]
````
